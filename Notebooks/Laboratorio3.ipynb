```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

```python
# Semilla para reproducibilidad
np.random.seed(42)

# Listas de datos simulados
influencers = ['@cancun_traveler', '@playa_dreams', '@mexico_lifestyle', '@beach_paradise', '@yucatan_explorer']
tipos_post = ['foto', 'reel', 'carrusel']

# Rango de fechas (últimos 6 meses)
end_date = datetime(2026, 5, 29)
start_date = end_date - timedelta(days=180)
```

```python
num_registros = 100000

data = {
    'influencer': np.random.choice(influencers, num_registros),
    'likes': np.random.randint(500, 250000, num_registros),
    'comentarios': np.random.randint(0, 15000, num_registros),
    'tipo_post': np.random.choice(tipos_post, num_registros, p=[0.5, 0.3, 0.2]),  # 50% foto, 30% reel, 20% carrusel
    'fecha': [start_date + timedelta(days=np.random.randint(0, 180)) for _ in range(num_registros)]
}

df = pd.DataFrame(data)
```

```python
# Crear columna de engagement (likes + comentarios)
df['engagement'] = df['likes'] + df['comentarios']

# Engagement promedio por tipo de post
engagement_por_tipo = df.groupby('tipo_post')['engagement'].mean().sort_values(ascending=False)
print("Engagement promedio por tipo de post:")
print(engagement_por_tipo)
```

```python
# Top 3 influencers con más likes acumulados
top_influencers = df.groupby('influencer')['likes'].sum().sort_values(ascending=False).head(3)
print("\nTop 3 influencers por likes totales:")
print(top_influencers)

# Mejor día de la semana para publicar (si tienes fecha)
df['dia_semana'] = pd.to_datetime(df['fecha']).dt.day_name()
mejor_dia = df.groupby('dia_semana')['engagement'].mean().sort_values(ascending=False)
print("\nEngagement promedio por día de la semana:")
print(mejor_dia)
```
